```python
import numpy as np
EMG_SANO = np.loadtxt("EMG_Sano.txt")

Tiempo_EMG_SANO = EMG_SANO[:,0]
Magnitud_EMG_SANO  = EMG_SANO[:,1]

```

```python
import matplotlib.pyplot as plt



# Calculo de varianza y desviación estándar
variance = np.var(Magnitud_EMG_SANO)
std_dev = np.std(Magnitud_EMG_SANO)

# graficar la señal
plt.plot(Tiempo_EMG_SANO, Magnitud_EMG_SANO, label='Señal EMG')

#Marcar la varianza
plt.fill_between(Tiempo_EMG_SANO, Magnitud_EMG_SANO - variance, Magnitud_EMG_SANO + variance, alpha=0.2, label='variance')

#Marcar la desviación estándar
#plt.errorbar(Tiempo_EMG_SANO, Magnitud_EMG_SANO, yerr=std_dev, linestyle='None', marker='o', label='standard deviation')

# Etiquetas
plt.legend()
plt.xlabel('Tiempo (S)')
plt.ylabel('Magnitud (uV)')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
EMG_SANO = np.loadtxt("EMG_Sano.txt")
# Cargar la info y cortar el vector a las primeras 1000 muestras
emg_data = EMG_SANO[1001:3001,1]

# Calcular varianza y desviación estándar
variance = np.var(emg_data)
std_dev = np.std(emg_data)

#Agrandar la imagen
plt.figure(figsize=(22, 18))


#Graficar la ingo
plt.plot(emg_data, label='Señal EMG')

#Linea de varianza
plt.axhline(y=variance, color='r', linestyle='--', label='Varianza')

# Sobreado de desviación estándar
plt.fill_between(range(len(emg_data)), emg_data - std_dev, emg_data + std_dev, alpha=0.3, label='Desviación estándar')

# legendas y etiquetado
plt.legend()
plt.xlabel('                  Muestras                  Plot EMG, Biomédica, Ing Rafael Castro')
plt.ylabel('                Amplitud EMG              Plot EMG, Biomédica, Ing Rafael Castro')

# imprimer la grafica
plt.show()



```
